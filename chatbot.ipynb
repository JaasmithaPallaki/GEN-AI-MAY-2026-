```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('Gemini'))

knowledge_base = """
COLLEGE: VVIT Engineering College

==== ADMISSIONS ====
- Q: How can I apply for admission?
  A: You can apply online through the college admission portal by filling out
     the application form and uploading required documents.

- Q: What are the courses available?
  A: We offer B.Tech, M.Tech, MBA, MCA, and Diploma courses in multiple branches.

- Q: What is the admission eligibility for B.Tech?
  A: Students must complete 12th grade with Physics, Chemistry, and Mathematics.

==== FEES ====
- Q: How can I pay the college fees?
  A: Fees can be paid online through UPI, Debit/Credit Card, or Net Banking.

- Q: Is there any scholarship available?
  A: Yes, merit-based and government scholarships are available for eligible students and based on college entrance test.

==== HOSTEL ====
- Q: Does the college provide hostel facilities?
  A: Yes, separate hostel facilities are available for boys and girls.

- Q: What facilities are available in the hostel?
  A: Wi-Fi, mess, study rooms, laundry, and 24/7 security are available.

==== EXAMINATION ====
- Q: How can I check exam results?
  A: Results are published on the college student portal.

- Q: What happens if I fail in a subject?
  A: You can apply for supplementary examinations.

==== LIBRARY ====
- Q: What are the library timings?
  A: The library is open from 8:00 AM to 8:00 PM on working days,
     on first and third Saturday it is from 9:00 AM to 5:00 PM,
     and Sunday is a holiday.

==== PLACEMENTS ====
- Q: Does the college provide placements?
  A: Yes, top companies visit the campus every year for placements.

CONTACT:
Website: www.vvitcollege.edu
Phone: 9876543210
"""

system_instruction = f"""
You are a helpful college support chatbot.

Rules:
1. Answer ONLY using the knowledge base.
2. If information is unavailable, reply:
   "I'm sorry, I don't have that information. Please contact the college office."
3. Keep answers short, polite, and clear.
4. Reply in 1-3 sentences only.

==== KNOWLEDGE BASE ====
{knowledge_base}
"""

model = genai.GenerativeModel(
    model_name='gemini-2.5-flash-lite',
    system_instruction=system_instruction
)

chat = model.start_chat(history=[])

print("Hi! I am the College FAQ Assistant.")
print("Type quit, bye, or exit to stop.\n")

while True:

    user_input = input("You: ")


    if user_input.lower() in ["quit", "bye", "exit"]:
        print("Bot: Goodbye!")
        break

    if user_input.lower() in ["thank you", "thanks", "thankyou"]:
        print("Bot: You're welcome!")
        break

    reply = chat.send_message(
        user_input,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 300
        }
    )

    print(f"College Bot: {reply.text}\n")
```
