```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('Gemini'))
```

```python
def generate_blog(topic, tone, length_words=500):

    system_prompt = (
        "You are an expert blog writer. "
        "You creats a well-structured blog posts in a clear human tone."
    )

    user_prompt = f"""
    Write a blog post.

    Topic: {topic}
    Tone: {tone}
    Length: approximately {length_words} words
    """

    model = genai.GenerativeModel(
        model_name="gemini-2.5-flash-lite",
        system_instruction=system_prompt
    )


    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 700
        }
    )

    return response.text
```

```python
def generate_blog(topic, tone, length_words=500):

    system_prompt = (
        "You are a professional blog writer. "
        "You create engaging, well-structured, and easy-to-read blog posts in a natural human tone."
    )

    user_prompt = f"""
    Write a blog post.

    Topic: {topic}
    Tone: {tone}
    Length: approximately {length_words} words
    """

    model = genai.GenerativeModel(
        model_name="gemini-2.5-flash",
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 700
        }
    )

    return response.text



blog = generate_blog(
    topic="Benefits of Artificial Intelligence",
    tone="Friendly",
    length_words=300
)

print(blog)
```

```python
topic = input("Enter Blog Topic: ")
tone = input("Enter Blog Tone: ")
length = int(input("Enter Blog Length in Words: "))

print("\nGenerating Blog...\n")


blog = generate_blog(topic, tone, length)
print(blog)
```

```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get('Gemini'))


def generate_email(recipient, purpose, tone):

    system_prompt = (
        "You are an expert email assistant. "
        "You write professional and well-structured emails "
        "with proper greetings, body content, and closing statements."
    )

    user_prompt = f"""
    Recipient Name: {recipient}
    Email Purpose: {purpose}
    Writing Tone: {tone}
    """

    model = genai.GenerativeModel(
        model_name='gemini-2.5-flash-lite',
        system_instruction=system_prompt
    )

    response = model.generate_content(
        user_prompt,
        generation_config={
            "temperature": 0.7,
            "max_output_tokens": 300
        }
    )

    return response.text



email = generate_email(
    recipient="HR Manager",
    purpose="Requesting leave for two days",
    tone="Professional"
)


print(email)
```

```python
recipient = input("Enter Recipient Name: ")
purpose = input("Enter Email Purpose: ")
tone = input("Enter Email Tone: ")


print(generate_email(recipient, purpose, tone))
```
